## Section 3.1 - SIFT applied for feature detection

We want to represent the visual features as a fixed-length vectors and match them efficiently to detect loop closures. The raw descriptors are assigned to the nearest visual word in the vocabulary, and a histogram of visual words is constructed for each image or point cloud. This histogram represents the BoW vector for the image, which is a fixed-size representation of the image features. This BoW vector can then be used for efficient matching and loop closure detection.
